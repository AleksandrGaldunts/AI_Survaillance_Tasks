```python
import cv2
import numpy as np
from IPython.display import display, Image

# Convert img from BGR to RGB for compatibility
img_rgb = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)

# Concatenate images
img_concats = np.concatenate((img_rgb, img_matplotlib), axis=1)

# Encode image as PNG in memory
_, img_encoded = cv2.imencode('.png', cv2.cvtColor(img_concats, cv2.COLOR_RGB2BGR))

# Display image in Jupyter Notebook
display(Image(data=img_encoded.tobytes()))
```
